Import python libraries required for the script

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import warnings

from mpl_toolkits.mplot3d import Axes3D
from scipy import stats
from scipy.stats import norm
from scipy.stats import linregress
from sklearn.preprocessing import StandardScaler
from numpy.polynomial import polynomial as npp

warnings.filterwarnings('ignore')

%matplotlib inline 
```

Read the CSV files

```python
df_pa = pd.read_csv('VMI_Data_BP1_V02_AK6_mod02.csv')


# Limiting only 10000 randomly selected rows for quick rendering of the charts
df_pa=df_pa.sample(10000)
```

Check the columns in the data

```python
print(df_pa.columns)
```

```python
%matplotlib inline  
%matplotlib notebook

# Splice Delta x Cut Length x Wait Time, or Splice Delta x Moving Average5 x Moving Average10
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')

for so, col in enumerate(['Bad', 'OK', 'Good']):
#     print(so, col)
    xs = df_pa[(df_pa['SPLICE_GRADE']==col)].LFT_SPLICE_DELTA
    ys = df_pa[(df_pa['SPLICE_GRADE']==col)].MID_SPLICE_DELTA
    zs = df_pa[(df_pa['SPLICE_GRADE']==col)].RHT_SPLICE_DELTA
#     sov = df_pa[(df_pa['SPLICE_GRADE']==so)].SPLICE_GRADE
    if col == 'Bad':
        colr = 'Orange'
    elif col == 'OK':
        colr = 'Blue'
    else:
        colr = 'Green'
    ax.scatter(xs, ys, zs, c=colr, s=50, alpha=0.6, edgecolors='w', label=col)

ax.set_xlabel('Left Splice Delta')
ax.set_ylabel('Mid Splice Delta')
ax.set_zlabel('Right Splice Delta')

# ax.set_xlim(-20, 20)
# ax.set_ylim(-20, 20)
# ax.set_zlim(-20, 20)

# ax.mouse_init()

ax.legend()

plt.show()
```
